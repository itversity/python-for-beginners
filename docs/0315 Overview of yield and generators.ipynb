* Getting Started with generators

````python
import time
def process(l):
    ls = []
    for i in l:
        time.sleep(1)
        ls.append(i * i)
    return ls
````

````python
l = [1, 3, 2, 5, 8]
````

````python
process(l)
````

````python
import time
def process_list(l):
    for i in l:
        time.sleep(1)
        yield i * i
````

````python
l = [1, 3, 2, 5, 8]
````

````python
process_list(l)
````

````python
for i in process_list(l):
    print(i)
````

* Overview of developing custom generators

Here is an example for custom generator `read_lines`. We have not covered File I/O yet. We will see the implementation and validation of this generator as part of lectures related to File I/O.
```python
def read_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.rstrip('\n')

for line in read_lines('sales.csv'):
    print(line)
```
