```python
# Create table in Database using Python
import psycopg2
```

```python
conn = psycopg2.connect(
    host='localhost',
    port=5432,
    database='itversity_retail_db',
    user='itversity_retail_user',
    password='itversity'
)
```

```python
cur = conn.cursor()
```

```python
cur.execute('DROP TABLE IF EXISTS departments')
```

```python
cur.execute('''
    CREATE TABLE departments (
        department_id INT PRIMARY KEY,
        department_name VARCHAR(30)
    )
''')
```

```python
cur.execute('SELECT * FROM departments')
```

```python
cur.fetchall()
```

```python
# Read CSV Data from files
with open('data/retail_db/departments/part-00000') as fp:
    data = fp.read().splitlines()
```

```python
# Data Type Conversion of non-string columns

departments = []
for rec in data:
    r = rec.split(',')
    departments.append((int(r[0]), r[1]))

departments
```

```python
# Populate Data into Table
query = '''
    INSERT INTO departments (department_id, department_name)
    VALUES (%s, %s)
'''
```

```python
cur.executemany(query, departments)
```

```python
conn.commit()
```

```python
# Validate Data in the table
cur.execute('SELECT * FROM departments')
```

```python
cur.fetchall()
```

```python
for rec in cur:
    print(rec)
```

```python
import psycopg2
```

```python
cur = conn.cursor()
```

```python
cur.execute('DROP TABLE IF EXISTS sales')
```

```python
cur.execute('''
    CREATE TABLE sales (
        sale_id INT PRIMARY KEY,
        sale_rep_id INT,
        sale_amount FLOAT,
        commission_pct INT
    )
''')
```

```python
with open('data/sales/part-00000') as fp:
    data = fp.read().splitlines()
```

```python
type(data)
```

```python
recs = []
for rec in data[1:]:
    r = rec.split(',')
    recs.append((int(r[0]), int(r[1]), float(r[2]), int(r[3]) if r[3] != '' else None))

recs
```

```python
query = '''
    INSERT INTO sales (sale_id, sale_rep_id, sale_amount, commission_pct)
    VALUES (%s, %s, %s, %s)
'''
```

```python
cur.executemany(query, recs)
```

```python
conn.commit()
```
